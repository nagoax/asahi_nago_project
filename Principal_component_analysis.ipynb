# Prepare to perform PCA

```python
#import StandardScaler to perform  standarization
from sklearn.preprocessing import StandardScaler

#項目「打率」以降の標準化
# Standarize all data after AVG column. 
# Standardization is the process of putting different variables on the same scale. This process allows you to compare scores between different types of variables.
# Calculating z-score for columns after AVG. Create array.
df2 = df.loc[:,'AVG':].astype('float')
df2
sc = StandardScaler()
sc_df = sc.fit_transform(df2)
sc_df
```

```python
from sklearn.decomposition import PCA
#Create models to perform PCA
model = PCA(n_components=2, whiten=True)
#Model training using the standardrized data
model.fit(sc_df)
```

```python
#Transform standardrized data into list, and change into dataframe.
new = model.transform(sc_df)
new_df = pd.DataFrame(new)
print(new_df.head())
print(new_df.tail())
new_df.head()
```

```python
#Compute the component(data that describes the feature for all columns )
#Displays the first trained component.
print(model.components_[0])
```

```python
#Create another training model
new = model.transform(sc_df)
new_df = pd.DataFrame(new)
new_df.head(5)
```

```python
#Create new columns called PC1 and PC2
new_df.columns=['PC1','PC2'] #生成された軸に仮の名前を付けておく

#標準化データと生成された2軸を合体
#Join standardrized data and training model.
df3 = pd.DataFrame(sc_df,columns=df2.columns)
df4 = pd.concat([df3,new_df],axis=1)
#打率～長打率とPC1の相関性を確認
#Check the correlation between all columns and PC1
#We will rename this PC to Slugger later. These components will determine how the players good at hitting extra-base hitting or HR.
df_corr = df4.corr()
pc_corr = df_corr.loc[:'SLG','PC1':]
pc_corr['PC1'].sort_values(ascending=False)
```

```python
#打率～長打率とPC2の相関性を確認
#Do the same perform to PC2
#Rename as Speedstar. Speedstar determines how players are fast-runner.
pc_corr['PC2'].sort_values(ascending=False)
```

# PCA 

```python
#PCの名前変更
#Rename PC to Slugger and Speedstar
col = ['Slugger','Speedstar']
new_df.columns = col

#選手名を追加
#Bring back the Name and Team columns and join.
df_name = df[['Name','Team']]
df5 = pd.concat([df_name,new_df],axis=1)

df5.head()
```

```python
import matplotlib.pyplot as plt
x = df5['Slugger']
y = df5['Speedstar']
labels = df5['Name'] #データラベル用
#Plotting all data with x and y
plt.figure(figsize=(36,24))
plt.xlabel('Slugger', fontsize=24)
plt.ylabel('Speedstar', fontsize=24)
plt.scatter(x, y, s=300)
for i, label in enumerate(labels):
    plt.text(x[i], y[i], label, fontsize=24)
plt.grid(True)
plt.show()
```

There is a player who has high correlation in Sluuger. As you can see fron his data, he has high score in HR, and TB. The PCA worked well using the data from NPB qualified batter in 2022.
