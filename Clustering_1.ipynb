```python

#データの標準化
#import StandardScaler to perform standarization
from sklearn.preprocessing import StandardScaler
# Standarize all data after AVG column. 
# Standardization is the process of putting different variables on the same scale. This process allows you to compare scores between different types of variables.
# Calculating z-score for columns after AVG. Create array.
df3 = df2.loc[:,'AVG':].astype('float')
sc = StandardScaler()
sc_df = sc.fit_transform(df3)

from sklearn.decomposition import PCA
#Create models to perform PCA
#モデル作成
model = PCA(n_components=2, whiten=True)
#Model training using the standardrized data
#モデル学習
model.fit(sc_df)

#Transform standardrized data into list, and change into dataframe.
#作成した主成分でデータフレーム作成
new = model.transform(sc_df)
new_df = pd.DataFrame(new)
new_df.columns=['PC1','PC2'] #主成分に仮命名
new_df.head(5)
```
